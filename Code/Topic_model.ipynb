```python
import pandas as pd
labels = ["Nhắc nhở", #0
"Tâm sự tình cảm","Mua bán tài liệu", #1
"Tìm đồ", "Tư vấn", #3
"Chia sẻ trải nghiệm","Chia sẻ kiến thức", #5
"Sự kiện trường","Ngày lễ", #7
"Hỏi-đáp","Cảnh báo", #9
"Phản ánh, bức xúc","Vấn đề học tập", #11
"Vấn đề trường","Tuyển dụng",  #13
"Khác"] #15
df2 = pd.read_csv("./df2.csv")
data_augmented = pd.read_csv("./data_augmented_p.csv")
df_train = pd.concat((df2, data_augmented))
for label in labels:
    print(len(df_train[df_train["Label"]==label]),": " + label)
print(len(df_train[pd.isna(df_train["Label"])]), ": Null")
```

```python
def preprocess(df) -> pd.DataFrame:
    df_train = df.copy()
    df_train = df_train[~df_train["Post_Content"].str.contains("tiệm")]
    df_train = df_train[pd.notna(df_train["Label"])]
    # df_train = df_train[df_train["Label"]!="Khác"]
    # make "Tư vấn" equal to "Hỏi-đáp"
    df_train["Label"] = df_train["Label"].replace("Tư vấn", "Hỏi-đáp")
    df_train["Label"] = df_train["Label"].replace("Nhắc nhở", "Chia sẻ kiến thức")
    df_train["Label"] = df_train["Label"].replace("Vấn đề học tập", "Vấn đề trường")
    # df_r["Label"] = df_r["Label"].replace("Sự kiện trường", "Vấn đề trường")
    # df_r = df_r.drop(df_r[df_r["Label"]=="Mua bán tài liệu"][0:600].index)
    # df_r = df_r.drop(df_r[df_r["Label"]=="Hỏi-đáp"][0:200].index)
    return df_train
df2 = preprocess(df2)
data_augmented = preprocess(data_augmented)
df_train = pd.concat((df2, data_augmented))
for label in labels:
    print(len(df_train[df_train["Label"]==label]),": " + label)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('keepitreal/vietnamese-sbert')
sentences_main = df2["Post_Content"].to_list()
sentences_aug = data_augmented["Post_Content"].to_list()
embeddings_main = model.encode(sentences_main)
embeddings_aug = model.encode(sentences_aug)
# embeddings = np.concatenate((embeddings_main,embeddings_aug))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(embeddings_main, df2["Label"], test_size=0.2)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
X_train_pre, X_test, y_train_pre, y_test = train_test_split(embeddings_main, df2["Label"], test_size=0.2)
X_train = np.concatenate((X_train_pre, embeddings_aug))
y_train = pd.concat((y_train_pre, data_augmented["Label"]))
```

```python
# labels = y_train.unique()
# label_dict = {}
# for i, label in enumerate(labels):
#     label_dict[label] = i
label_dict = {'Hỏi-đáp': 0,
'Tuyển dụng': 1,
'Khác': 2,
'Mua bán tài liệu': 3,
'Chia sẻ kiến thức': 4,
'Tìm đồ': 5,
'Ngày lễ': 6,
'Tâm sự tình cảm': 7,
'Chia sẻ trải nghiệm': 8,
'Vấn đề trường': 9,
'Phản ánh, bức xúc': 10,
'Cảnh báo': 11,
'Sự kiện trường': 12}
label2idx = {v: k for k, v in label_dict.items()}
```

```python
from keras.utils import to_categorical

#One hot encoding
encoded_y_train = y_train.map(label_dict)
encoded_y_test = y_test.map(label_dict)
y_train_cat=to_categorical(encoded_y_train,len(label2idx))
y_test_cat=to_categorical(encoded_y_test,len(label2idx))
```

## CNN

```python
import tensorflow as tf
from keras.losses import categorical_crossentropy
from keras.callbacks import ModelCheckpoint
# checkpoint
filepath = "./model.keras"
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
import keras
model = keras.Sequential([
    keras.layers.Conv1D(64,kernel_size=3,padding="same", input_shape=(768,1), activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(pool_size=2, strides=2),

    keras.layers.Conv1D(128,kernel_size=3,padding="same", activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(pool_size=2, strides=2),

    keras.layers.Conv1D(256,kernel_size=3,padding="same", activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(pool_size=2, strides=2),
    keras.layers.Conv1D(256,kernel_size=3,padding="same", activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(pool_size=2, strides=2),

    keras.layers.Conv1D(512,kernel_size=3,padding="same", activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(pool_size=2, strides=2),
    keras.layers.Conv1D(512,kernel_size=3,padding="same", activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(pool_size=2, strides=2),

    keras.layers.Conv1D(512,kernel_size=3,padding="same", activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(pool_size=2, strides=2),
    keras.layers.Conv1D(512,kernel_size=3,padding="same", activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPooling1D(pool_size=2, strides=2),

    keras.layers.Flatten(),
    keras.layers.Dense(1024, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(1024, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(len(label_dict), activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(len(label_dict), activation="softmax"),
])

model.compile(
    loss="categorical_crossentropy", # loss function here
    optimizer="adam",
    metrics=["accuracy"])

history = model.fit(X_train, y_train_cat, epochs=100, validation_data=[X_test, y_test_cat], verbose=1, callbacks=[checkpoint])
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input,Dense,Conv1D,Flatten,Embedding,MaxPool1D,concatenate,Dropout
from tensorflow.keras.callbacks import ModelCheckpoint,TensorBoard,EarlyStopping
from tensorflow.keras.optimizers import Adam

filepath = "./model2.keras"
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
layer_0 = Input(shape=(768,1))

embed_layer = layer_0

l_1 = Conv1D(32,4,activation="relu",kernel_initializer =tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(embed_layer)

l_2 = Conv1D(32,4,activation="relu",kernel_initializer =tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(embed_layer)

l_3 = Conv1D(32,4,activation="relu",kernel_initializer = tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(embed_layer)

second_layer = concatenate([l_1,l_2,l_3])

max_pool_1 = MaxPool1D(3)(second_layer)

l_4 = Conv1D(16,3,activation="relu",kernel_initializer = tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(max_pool_1)

l_5 = Conv1D(16,3,activation="relu",kernel_initializer = tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(max_pool_1)

l_6 =  Conv1D(16,3,activation="relu",kernel_initializer = tf.keras.initializers.he_normal(),kernel_regularizer=tf.keras.regularizers.l2())(max_pool_1)

third_layer = concatenate([l_4,l_5,l_6])

max_pool_2 = MaxPool1D(3)(third_layer)

fourth_layer = Conv1D(16,3,activation='relu',
                      kernel_initializer = tf.keras.initializers.he_normal(seed=42),kernel_regularizer=tf.keras.regularizers.l2())(max_pool_2)

flatten = Flatten()(fourth_layer)

dropout_layer = Dropout(0.2)(flatten)

dense_layer = Dense(32,activation="relu",kernel_initializer = tf.keras.initializers.he_normal())(dropout_layer)

output_layer = Dense(len(label_dict),activation="softmax",kernel_initializer= tf.keras.initializers.glorot_normal())(dense_layer)


cnn_model =Model(inputs=layer_0,outputs=output_layer)
history = model.fit(X_train, y_train_cat, epochs=100, validation_data=[X_test, y_test_cat], verbose=1, callbacks=[checkpoint])
```

```python
from keras.callbacks import ModelCheckpoint
# checkpoint
filepath = "./model_v2.3.keras"
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
import keras
model = keras.Sequential([

    keras.layers.Dense(768, input_shape=(768,), activation="relu"),
    # normalize
    keras.layers.BatchNormalization(),
    keras.layers.Dense(768, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(256, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(256, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(32, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(16, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(len(labels), activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(len(labels), activation="softmax"),
])

model.compile(
    loss="categorical_crossentropy", # loss function here
    optimizer="adam",
    metrics=["accuracy"])

history = model.fit(X_train, y_train_cat, epochs=100, validation_data=[X_test, y_test_cat], verbose=1, callbacks=[checkpoint])
```
